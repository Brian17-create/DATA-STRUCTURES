```python
# Number 1:

# a)
sentence = "Data Structures and Algorithms are important in problem solving"

       #  Count the number of words

word_count = len(sentence.split())

print("Number of words:", word_count)


# b)

#  Convert to list of words and display in reverse order

words = sentence.split()

print("Words in reverse order:", words[::-1])


# c)

#  Replace "important" with "essential"

new_sentence = sentence.replace("important", "essential")

print("New sentence:", new_sentence)



# Number 2:

# a)

#       i.         Big-O (O) - Upper bound of growth rate (worst-case performance).

#        ii.            Big-Θ (Θ) -Tight bound (both upper and lower) — average/exact performance.

#    iii.         Big-Ω (Ω) - Lower bound of growth rate (best-case performance)




# b)

# Time complexity: O(n^2)

# Efficiency: Not efficient for large n, because the number of operations grows quadratically.



# Number 3:

# Dummy array for search and sort functions
arr = [64, 34, 25, 12, 22, 11, 90, 32, 45]

# a)      Linear search
def linear_search(arr, target):
    for i in range(len(arr)):
        if arr[i] == target:
            return i
    return -1
print("Linear Search (32):", linear_search(arr, 32))




# b) Binary search
def binary_search(arr, target):
    left, right = 0, len(arr) - 1
    while left <= right:
        mid = (left + right) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1 # Corrected
    return -1

sorted_arr = sorted(arr)
print("Binary Search (45):", binary_search(sorted_arr, 45))



# c)Bubble sort
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
    return arr



print("Bubble Sort:", bubble_sort(list(arr)))





# Number 4:

# a)Node class

class Node:
    def __init__(self, data):
        self.data = data
        self.next = None


# b)

class LinkedList: # Corrected from "List:"
    def __init__(self):
        self.head = None

    # i. Insert at beginning

    def insert_beginning(self, data):
        new_node = Node(data)
        new_node.next = self.head
        self.head = new_node

# ii. Insert at end

    def insert_end(self, data):
        new_node = Node(data)
        if not self.head:
            self.head = new_node
            return
        temp = self.head
        while temp.next:
            temp = temp.next
        temp.next = new_node

# iii. Display all elements

    def display(self): # Corrected indentation
        temp = self.head
        while temp:
            print(temp.data, end=" -> ")
            temp = temp.next
        print("None")




# Number 5:

# a)

# Queue (FIFO) is most appropriate because students are served in the order they arrive.

#        Justification: Ensures fairness and models real-world waiting lines.

# b)

from collections import deque



class FinanceQueue:
    def __init__(self):
        self.queue = deque()



    # i. Add student

    def add_student(self, student):
        self.queue.append(student)
        print(f"{student} added to the queue.")

    # ii. Serve student

    def serve_student(self):
        if self.queue:
            served = self.queue.popleft()
            print(f"{served} has been served.")
        else:
            print("No students in queue.")



    # iii. Display queue

    def display_queue(self):
        print("Current queue:", list(self.queue))
```
